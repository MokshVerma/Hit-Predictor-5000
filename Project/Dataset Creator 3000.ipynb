```python
albumList = ['Hot Fuss', 'Hybrid Theory','Minutes To Midnight', 'Gore', 'Illuminate', 'Halcyon Days',
            'One More Light', 'Concrete And Gold', 'The Hunting Party', 'Born to Die', 'Konvicted', 
            'Freedom', 'A Head Full Of Dreams', 'Absolution', 'Favourite Worst Nightmare', 'PCD', 
            'Tha Blue Carpet Treatment', 'Shock Value', 'Results May Vary', 'Famous First Words', 
            'Nine Track Mind', 'Sam\'s Town', 'Wonderful Wonderful', 'Sawdust', 'Hamilton (Original Broadway Cast Recording)',
            'Song Cycle', 'Invincible', 'Metal Machine Music', 'Make A Shadow', 'Lust For Life', 
             'Last Young Renegades','Gone Now','Neva Left','Waiting on a Song','The Transition of Mali',
             'Something to Tell You','Talkie Walkie','Now Here is Nowhere','Musicology',
             'Genius Loves Company','Room on the 3rd Floor', 'Dark Sky Island', 'Dodge and Burn', 'Cherry Bomb', 
            'Postcards from Paradise', 'The Monsanto Years', 'Heart Blanche', 'Wallflower', 'Kindred', 'Title', 
             'Post-Nothing']




len(albumList)
```

```python
#Artist Popularity
#Artist Genre
#Artist Listeners
for i in albumList:
    album = spotify.search(q=i, type='album')
    if(len(album['albums']['items'])!=0):
        album_id = album['albums']['items'][0]['id']
        no_of_tracks = album['albums']['items'][0]['total_tracks']
        album_tracks = spotify.album_tracks(album_id)
        for j in range(0, no_of_tracks):
            track_id = album_tracks['items'][j]['id']
            popularity = spotify.track(track_id)
            popularityList.append(popularity['popularity'])
            print(track_id)
            songName = album_tracks['items'][j]['name']
            print(songName)
            songNameList.append(songName)

            artistName = album_tracks['items'][j]['artists'][0]['name']
            artistNameList.append(artistName)

            audio_features = spotify.audio_features(track_id)
            danceabilityList.append(audio_features[0]['danceability'])
            energyList.append(audio_features[0]['energy'])
            loudnessList.append(audio_features[0]['loudness'])
            speechinessList.append(audio_features[0]['speechiness'])
            acousticList.append(audio_features[0]['acousticness'])
            tempoList.append(audio_features[0]['tempo'])
            trackidList.append(audio_features[0]['id'])
            keyList.append(audio_features[0]['key'])
            instrumentalList.append(audio_features[0]['instrumentalness'])
            livenessList.append(audio_features[0]['liveness'])
            valenceList.append(audio_features[0]['valence'])
            durationList.append(audio_features[0]['duration_ms'])
            timeSignatureList.append(audio_features[0]['time_signature'])
            
            artists = spotify.search(q=artistName, type='artist')
            artistPopularity = artists['artists']['items'][0]['popularity']
            artistPopularityList.append(artistPopularity)
            artistFollowers = artists['artists']['items'][0]['followers']['total']
            artistFollowersList.append(artistFollowers)
            artistGenres = artists['artists']['items'][0]['genres']
            artistGenresList.append(artistGenres)

            lyricsQ = musixmatch.matcher_lyrics_get(songName, artistName)
            if(lyricsQ['message']['header']['status_code'] == 200):
                lyricsList.append(lyricsQ['message']['body']['lyrics']['lyrics_body'].
                                  strip(',').
                                  replace('\n', ' '))
            else:
                lyricsList.append('')
```

```python
df = pd.DataFrame({'Song Name': songNameList, 'Artist Name': artistNameList, 'Artist Popularity': artistPopularityList,
           'Artist Genres': artistGenresList, 'Artist Followers' : artistFollowersList, 'Danceability':danceabilityList, 
           'Energy':energyList, 'Track ID': trackidList, 'Loudness': loudnessList, 'Speechiness': speechinessList, 
           'Acousticness': acousticList, 'Tempo' :tempoList, 'Key': keyList, 'Instrumentalness': instrumentalList, 
           'Lyrics': lyricsList, 'Popularity': popularityList, 'Liveness': livenessList, 'Valence': valenceList, 
           'Time Signature List': timeSignatureList, 'Duration': durationList, 'Hit': hitList})
```
